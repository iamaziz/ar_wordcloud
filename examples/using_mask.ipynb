```python
from ar_wordcloud import ArabicWordCloud
from ar_wordcloud.utils import read_mask_image

heart_mask = 'https://user-images.githubusercontent.com/3298308/82387916-42d5a380-9a06-11ea-8dc3-ab493228858e.png' # image src: https://images.app.goo.gl/7JtyWX9HWe37c9ti9
awc = ArabicWordCloud(background_color="white", mask=read_mask_image(heart_mask), contour_width=4)
```

```python
awc.from_file('arabic.txt').to_image()
```

```python
# or with a title
awc.plot(awc.from_file('arabic.txt'), title='اللغة العربية جميلة')
```

```python
# save to a file
awc.from_file('arabic.txt').to_file('heart_mask.png');
```
